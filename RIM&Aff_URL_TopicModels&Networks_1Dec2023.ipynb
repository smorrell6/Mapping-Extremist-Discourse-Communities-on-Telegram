#Counting how many messages are in the network

## To count how many messages are in the network, I first iterate through the dataframe of messages and determine whether each message contains at least one of the Telegram URLs from the list of nodes in the network. I then count how many Telegram URLs appeared in the message (regardless of whether they are in the list of nodes). I count the messages that both contain at least one URL from the list of nodes and contain more than one Telegram URL as being part of the co-occurrence network. Because the co-occurrence network appears to only sometimes include messages which have repeats of the same URL, it is nessecary to have both of these components to establish that a message is in the co-occurence network.

```python
#creating a column for recording whether a message contains a URL that is in the list of nodes
df['contains_node_URL'] = ''
df.head()
```

```python
#going through each message to see if it contains a URL from the list of nodes
for row in range(len(df)):
  contains_node_URL = 0
  print(contains_node_URL)
  for URL in range(len(df_URLs)):
    if df_URLs['Id'][URL] in df['body'][row]:
      print("URL:" + df_URLs['Id'][URL])
      print("message:" + df['body'][row])
      contains_node_URL = 1
      print(contains_node_URL)
  df['contains_node_URL'][row] = contains_node_URL
```

#Add in network to see if the communities overlap with the topics

```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
#Import network from Gephi
df_network = pd.read_csv('RIM&Affil_Final4NovlNetwork_Edges.csv')
df_network.head()
```

```python
#Turn into a networkX graph
G = nx.from_pandas_edgelist(df_network, 'Source', 'Target', edge_attr='Id', create_using=None)
```

```python
#Assign communities using the Girvan-Newman Method
from networkx.algorithms.community.centrality import girvan_newman

communities_gn = girvan_newman(G)

num_communities_gn = 0

node_groups_gn = []
for com in next(communities_gn):
  node_groups_gn.append(list(com))
  num_communities_gn = num_communities_gn + 1

print(node_groups_gn)

print(num_communities_gn)

```

Add gn communities to the dataframe with the URLs and topics

```python
#Creating column for gn communities
df_URLs['gn_com'] = ''
```

```python
df_URLs.head()
```

```python
#Iterate through communities and add them to the df
#For each community add one and assign that number to the column - just use i as string
for i in range(len(node_groups_gn)):
  for URL in range(len(node_groups_gn[i])):
    for r in range(len(df_URLs)):
      if df_URLs['Id'][r] == node_groups_gn[i][URL]:
        df_URLs['gn_com'][r] = i
```
